```python
# Unique 'H1' values
h1_values = sorted(summary_df[summary_df['H1']<30]['H1'].unique())

# Initialize a dictionary to hold the frequency counts for each category
freq_counts = {i: [] for i in range(5)}

# Populate the dictionary with frequency counts for each 'H1' value
for i in range(5):
    temp_df = summary_df[(summary_df['variable'] == 'EP_POV') & (summary_df["percen_US"] == i) & (summary_df["H1"] > 0) & (summary_df["H1"] < 30)]
    value_counts = temp_df['H1'].value_counts().sort_index()
    for h1 in h1_values:
        freq_counts[i].append(value_counts.get(h1, 0))

# Plotting
bar_width = 0.15
index = range(len(h1_values))

fig, ax = plt.subplots(figsize=(20, 6))

# Plot each category as a grouped bar
for i in range(5):
    ax.bar([p + bar_width * i for p in index], freq_counts[i], bar_width, label=f'Overdose class {i}')

# Add labels and title
ax.set_xlabel('# of H1 persistence points')
ax.set_ylabel('Frequency')
ax.set_title('Frequency of H1 Persistence Points by Overdose Class - EP_POV')
ax.set_xticks([p + 2 * bar_width for p in index])
ax.set_xticklabels(h1_values)
ax.legend()

# Display the plot
plt.tight_layout()
plt.show()
plt.close()
```

### Check the sets

```python
var_counties ={}
```

```python
for var in variables:
    # print(var)

    # get the number of rows where the variable is equal to var and H1 is greater than 0
    temp = summary_df[(summary_df['variable'] == var) & (summary_df['H1'] > 0)]
    # print(len(temp))

    counties = temp['county'].unique()

    # add the number of counties to the dictionary
    var_counties[var] = counties

    
```

```python
# create a df with columns variable and counties
var_counties_df = pd.DataFrame(columns=['variable1', 'variable2','percentage'])
```

```python
# loop though the dictionary and match the counties
for i in range(variables.shape[0]):

    for j in range(variables.shape[0]):
        if i != j:
            print(variables[j])
            common_count = len(set(var_counties[variables[i]]).intersection(var_counties[variables[j]]))

            # print the percentage of counties that are common for the two variables
            print(f'Percentage of counties that are common for {variables[i]} is {common_count/len(var_counties[variables[i]])*100:.2f}%')
            print(f'Percentage of counties that are common for {variables[j]} is {common_count/len(var_counties[variables[j]])*100:.2f}%')

            #add the variables and the percentage to the df
            new_r1 = {'variable1':variables[i], 'variable2':variables[j], 'percentage':common_count/len(var_counties[variables[i]])*100}
            # new_r2 = {'variable1':variables[j], 'variable2':variables[i], 'percentage':common_count/len(var_counties[variables[j]])*100}
            
              # Use concat instead of append
            var_counties_df = pd.concat([var_counties_df, pd.DataFrame([new_r1])], ignore_index=True)
            # var_counties_df = pd.concat([var_counties_df, pd.DataFrame([new_r2])], ignore_index=True)


            print('\n')
            
```
